```python
# Parameter Size
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", size="size")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", size="size")
plt.show()
```

```python
# Parameter Hue
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", size="size", hue="size")
plt.show()
```

```python
# Parameter Alpha
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", size="size", hue="size", alpha=0.5)
plt.show()
```

```python
# Point Style
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", style="smoker", hue="smoker")
plt.show()
```
