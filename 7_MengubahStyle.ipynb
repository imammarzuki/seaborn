```python
# Mengubah Background
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mengatur gaya visual seaborn menjadi "dark"
sns.set_style("dark")

# Contoh dataset sederhana dengan seaborn
tips = sns.load_dataset("tips")

# Membuat box plot
sns.boxplot(x="day", y="total_bill", data=tips)

# Menambahkan label dan judul
plt.xlabel("Hari")
plt.ylabel("Total Tagihan")
plt.title("Distribusi Total Tagihan per Hari di Restoran")

# Menggunakan fungsi seaborn untuk menghilangkan garis atas dan kanan
sns.despine()

# Menampilkan plot
plt.show()
```

```python
# Mengubah Palette
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mengatur gaya visual seaborn menjadi "dark"
sns.set_style("dark")

# Mengatur Warna Element Pada Chart
sns.set_palette("PRGn_r")

# Contoh dataset sederhana dengan seaborn
tips = sns.load_dataset("tips")

# Membuat box plot
sns.boxplot(x="day", y="total_bill", data=tips)

# Menambahkan label dan judul
plt.xlabel("Hari")
plt.ylabel("Total Tagihan")
plt.title("Distribusi Total Tagihan per Hari di Restoran")

# Menggunakan fungsi seaborn untuk menghilangkan garis atas dan kanan
sns.despine()

# Menampilkan plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Menggunakan dataset bawaan seaborn
tips = sns.load_dataset("tips")

# Mengubah Scale yang Paling Besar
sns.set_context("poster")

# Membuat scatter plot dengan `relplot`
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", col="time", kind="scatter")

# Menampilkan plot
plt.show()
```
