```python
import seaborn as sns
import matplotlib.pyplot as plt

# Menggunakan dataset bawaan seaborn
tips = sns.load_dataset("tips")
tips["time"]

```

```python
# Membuat scatter plot dengan `relplot`
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", col="time", kind="scatter")

# Menampilkan plot
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="smoker", row="time")
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="day", col_wrap=2, col_order=["Sun", "Thur", "Fri", "Sat"])
plt.show()
```
