```python
import seaborn as sns
import matplotlib.pyplot as plt

tinggi = [162, 164, 169, 175, 166, 168, 165, 171, 176, 173]
berat = [54, 57, 60, 62, 55, 62, 66, 80, 77, 72]

sns.scatterplot(x=tinggi, y=berat)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

jnsklm = ["wanita", "Pria", "Wanita", "Wanita", "Wanita", "Wanita", "Pria", "Pria", "Pria", "Pria"]

sns.countplot(x=jnsklm, hue=jnsklm, palette="bright", legend=False)
plt.show()
```
